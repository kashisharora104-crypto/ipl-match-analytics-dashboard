```python
import pandas as pd

matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv.zip')

matches.head()
```

```python
mi_csk = matches[
    ((matches['team1']=='Mumbai Indians') & (matches['team2']=='Chennai Super Kings')) |
    ((matches['team1']=='Chennai Super Kings') & (matches['team2']=='Mumbai Indians'))
]

mi_csk[['id','season','date','team1','team2','winner']].sort_values('date', ascending=False)
```

```python
runs_per_over = match_df.groupby(['inning','over'])['total_runs'].sum().groupby(level=0).cumsum()

runs_per_over.unstack(level=0).plot(figsize=(10,5))
plt.title("MI vs CSK Score Progression")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.show()
```

```python
powerplay = match_df[match_df['over']<=6]
death = match_df[match_df['over']>=16]

print("Powerplay runs:", powerplay['total_runs'].sum())
print("Death overs runs:", death['total_runs'].sum())
```

```python
player_stats = match_df.groupby('batter').agg({
    'batsman_runs':'sum',
    'ball':'count'
})

player_stats['strike_rate'] = (player_stats['batsman_runs']/player_stats['ball'])*100

player_stats.sort_values('batsman_runs', ascending=False).head(5)
```

```python
runs_over = match_df.groupby(['inning','over'])['total_runs'].sum()

runs_over.unstack(level=0).plot(kind='bar', figsize=(12,5))
plt.title("Runs per Over (Manhattan Chart)")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.show()
```

```python
partnership_runs = match_df.groupby(['batter','non_striker'])['total_runs'].sum().sort_values(ascending=False).head(10)

partnership_runs
```

```python
def phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'

match_df['phase'] = match_df['over'].apply(phase)

bowling = match_df.groupby(['bowler','phase'])['total_runs'].sum().reset_index()

bowling
```

```python
cumulative_runs = match_df.groupby(['inning','over'])['total_runs'].sum().groupby(level=0).cumsum().reset_index()
```

```python
inning1 = cumulative_runs[cumulative_runs['inning']==1]
inning2 = cumulative_runs[cumulative_runs['inning']==2]
```

```python
import numpy as np

inning2['target'] = inning1['total_runs'].max()

inning2['runs_remaining'] = inning2['target'] - inning2['total_runs']

inning2['overs_remaining'] = 20 - inning2['over']

inning2['req_run_rate'] = inning2['runs_remaining'] / inning2['overs_remaining']

inning2['win_probability'] = 100 - (inning2['req_run_rate'] * 5)
```

```python
plt.figure(figsize=(10,5))
plt.plot(inning2['over'], inning2['win_probability'])
plt.title("Win Probability Curve")
plt.xlabel("Over")
plt.ylabel("Win %")
plt.show()
```
